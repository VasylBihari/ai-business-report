```python
from datetime import datetime

def parse_dates(x):
    for fmt in ("%d.%m.%Y", "%m/%d/%Y"):
        try:
            return datetime.strptime(x, fmt)
        except:
            continue
    return pd.NaT

df['Ship Mode'] = df['Order Date'].apply(parse_dates)
```

```python
from datetime import datetime

def parse_dates(x):
    for fmt in ("%d.%m.%Y", "%m/%d/%Y"):
        try:
            return datetime.strptime(x, fmt)
        except:
            continue
    return pd.NaT

df['Ship Date'] = df['Ship Date'].apply(parse_dates)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace(r'[^\w\s]', '', regex=True)
```

```python
df.drop(
    ['row_id', 'order_id', 'customer_name', 'country', 'postal_code', 'product_name',  ],
    axis=1,
    inplace=True
)
```

```python
df['discount'] = df['discount'].str.replace(',', '.', regex=False)
df['profit'] = df['profit'].str.replace(',', '.', regex=False)
df['sales'] = df['sales'].str.replace(',', '.', regex=False)
```

```python
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
```
